```python
data_vn = pd.read_csv('ToanQuoc.csv')
```

```python
import re

def parseline(j, toan, van, li, hoa, sinh, khtn, su, dia, gdcd, khxh, anh):
    res = re.findall('\d+\.?\d*',data_vn['DIEM_THI'][j])
    match = re.search(r'Toán:', data_vn['DIEM_THI'][j])
    i = 0
    if match:
        toan.append(res[i])
        i=i+1
    match = re.search(r'Ngữ văn:', data_vn['DIEM_THI'][j])
    if match:
        van.append(res[i])
        i=i+1
    match = re.search(r'Vật lí:', data_vn['DIEM_THI'][j])
    if match:
        li.append(res[i])
        i=i+1
    match = re.search(r'Hóa học:', data_vn['DIEM_THI'][j])
    if match:
        hoa.append(res[i])
        i=i+1
    match = re.search(r'Sinh học:', data_vn['DIEM_THI'][j])
    if match:
        sinh.append(res[i])
        i=i+1
    match = re.search(r'KHTN:', data_vn['DIEM_THI'][j])
    if match:
        khtn.append(res[i])
        i=i+1
    match = re.search(r'Lịch sử:', data_vn['DIEM_THI'][j])
    if match:
        su.append(res[i])
        i=i+1
    match = re.search(r'Địa lí:', data_vn['DIEM_THI'][j])
    if match:
        dia.append(res[i])
        i=i+1
    match = re.search(r'GDCD:', data_vn['DIEM_THI'][j])
    if match:
        gdcd.append(res[i])
        i=i+1
    match = re.search(r'KHXH:', data_vn['DIEM_THI'][j])
    if match:
        khxh.append(res[i])
        i=i+1
    match = re.search(r'Tiếng Anh:', data_vn['DIEM_THI'][j])
    if match:
        anh.append(res[i])
```

```python
toan = []
li = []
hoa = []
sinh = []
van = []
su = []
dia = []
gdcd = []
khtn = []
khxh = []
anh = []
for j in range(len(data_vn['DIEM_THI'])):
    if pd.isnull(data_vn['DIEM_THI'][j]):
        continue
    else:
        parseline(j, toan, van, li, hoa, sinh, khtn, su, dia, gdcd, khxh, anh)
```
